```python
from sklearn.datasets import fetch_20newsgroups
import pandas as pd 

# Load train subset
data = fetch_20newsgroups(
    subset='train',
    remove=('headers', 'footers', 'quotes')  # remove noisy text
)

# Use first 5000 samples (CPU-friendly)
x = data.data[:5000]
y = data.target[:5000]

print("Number of samples:", len(x))
print("Number of classes:", len(set(y)))
```

```python
df = pd.DataFrame({'text': x, 'label': y})
df.head()
```
